```python
16.
topologicalSort(Graph G):
    // 1. 모든 정점의 진입 차수(in-degree) 계산
    for each vertex v in G:
        in_degree[v] = 0
    for each vertex v in G:
        for each neighbor u of v:
            in_degree[u]++

    // 2. 진입 차수가 0인 모든 정점을 Stack에 push
    S = empty stack
    for each vertex v in G:
        if in_degree[v] == 0:
            S.push(v)

    // 3. 위상 정렬 수행
    result = empty list
    while S is not empty:
        v = S.pop()
        result.append(v)

        // 방문한 정점에서 나가는 간선 제거 (이웃 정점의 진입 차수 감소)
        for each neighbor u of v:
            in_degree[u]--
            // 새롭게 진입 차수가 0이 된 정점을 Stack에 push
            if in_degree[u] == 0:
                S.push(u)

    // 4. 결과 출력 (사이클이 없다면 result의 크기는 정점의 총 개수와 같음)
    return result
```
